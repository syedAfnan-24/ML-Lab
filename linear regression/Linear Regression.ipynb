```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Food-Truck(For Linear Regression Program).csv',header=None)
```

```python
y=df[1]
```

```python
x = df[0]
```

```python
import numpy as np
import math
```

```python
xm=x.mean()
```

```python
xm
```

```python
ym = y.mean()
```

```python
ym
```

```python
num=0
denom=0
for i in range(len(x)):
  num=num+(x[i]-xm)*(y[i]-ym)
  denom = denom+(x[i]-xm)**2
```

```python
corr_num = (len(x)*((x*y).sum())) - (x.sum()*y.sum())
corr_denom = np.sqrt((len(x)*((x*x).sum())-(x.sum())**2)*((len(x)*(y*y).sum())-(y.sum())**2))
correlation = corr_num/corr_denom
slope = correlation * (num/denom)
intercept = ym - (slope*xm)
```

```python
print(slope)
intercept
```

```python
yp=[]
res=0
for i in range(len(x)):
  res = (x[i]*slope)+intercept
  yp.append(res)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.plot(x,yp)
plt.show()
```

```python
ssr=0
for i in yp:
  ssr=ssr+(i-ym)**2
ssr
```
